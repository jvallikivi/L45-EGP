```python
import torch
import matplotlib.pyplot as plt
import numpy as np
from glob import glob

def get_test_score(file_name):
    data = torch.load(file_name)
    return data['Test']

def show_curves(file_name):
    data = torch.load(file_name)
    fig, axs = plt.subplots(ncols=2)
    fig.set_size_inches(10, 5)
    axs[0].plot(data['train_curve'])
    axs[0].set_title("Training set")
    axs[1].plot(data['valid_curve'])
    axs[1].set_title("Validation set")

def print_args(file_name):
    data = torch.load(file_name)
    print(data['args'])

all_file_names = {}
```

# GIN

```python
scores = []
file_names = sorted(glob("./results/ogbg_molhiv_gin_[0-9]*.pt"))
all_file_names['GIN'] = file_names
for file_name in file_names:
    scores.append(get_test_score(file_name))
    print_args(file_name)
print(f"Performance: {np.mean(scores):.4f} Â± {np.std(scores):.4f}")
print([float(f"{score:.4f}") for score in scores])
show_curves(file_names[0])
```

```python
model_names = []
fig = plt.figure()

fig.set_size_inches(12, 7)
for i, (k, file_names) in enumerate(all_file_names.items()):
    model_names.append(k)
    scores = []
    for file_name in file_names:
        scores.append(get_test_score(file_name))
    mean = np.mean(scores)
    std = np.std(scores)
    sc = plt.scatter([i], [mean])
    col = sc.get_facecolors()[0].tolist()
    plt.scatter([i+0.03 for _ in scores], scores, color=col, alpha=0.2, s=5)
    plt.errorbar(i,mean,yerr=std, color=col, linestyle="None", solid_capstyle='projecting', capsize=5)
plt.xticks(np.arange(len(model_names)), model_names);
```
